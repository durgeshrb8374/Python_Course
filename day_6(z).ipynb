```python
#given three integers between 1 and 11
#if there sum is less than or == 21;return sum
#if sum exceeds 21 and there an 11,reduce by 10
#finally if sum(even after adj) >21 return burst

def fun(n1,n2,n3):
    s=n1+n2+n3
    if s<=21:
        return s
    elif s >21 and 11 in [n1,n2,n3]:
        s=s-10
        if s <= 21:
            return s
        else:
            return 'Bust'
    elif s>21:
        return 'Bust'
```

```python
fun(9,9,11)
fun(9,9,9)
fun(5,2,7)
```

```python
#switching of values and keys from d1 and d2
def switch(d1,d2):
    dout={}
    for d1key,d2val in zip(d1,d2.values()):
        dout[d1key]=d2val
    return dout
    
```

```python
#factorial using recursion
def fact(n):
    if n==1 or n==0:
        return 1
    else:
        return n*fact(n-1)
```

```python
fact(5)
```

```python
def fibo(n):
    if n==1:
        return 1
    elif n==2:
        return 1
    elif n>2:
        return fibo(n-1) + fibo(n-2)
    
```

```python
fibo(15)
```
