### Write a program that print the integer from 1 to 100
but for multiple of 3 print Fizz"
multiple of 5 print "Buzz"
bothh FizzBuzz

```python
for i in range(1,101):
    if (i%3==0) and (i%5==0):
        print('FizzBuzz')
    elif i%3==0:
        print('Fizz')
    
    elif i%5==0:
        print('Buzz')
    
    else:
        print(i)
```
